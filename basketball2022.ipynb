```python
import pandas as pd
```

### Web scraping data using pandas

```python
url = 'https://www.basketball-reference.com/leagues/NBA_2022_per_game.html'

html = pd.read_html(url, header=0)
df2022 = html[0]

df2022
```

```python
df = df2022.drop(df2022[df2022.Age == "Age"].index)

df
```
